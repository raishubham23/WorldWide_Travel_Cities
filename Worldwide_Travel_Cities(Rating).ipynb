# **Count of Cities by Country**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=WT_filtered, y='country', order=WT_filtered['country'].value_counts().index)
plt.title('Number of Cities by Country')
plt.xlabel('Number of Cities')
plt.ylabel('Country')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Count number of cities per country
country_counts = WT_filtered['country'].value_counts()

# Step 2: Get top 10 and bottom 10 countries
top_10 = country_counts.head(10)
bottom_10 = country_counts.tail(10)

# Step 3: Combine top and bottom into one Series
top_bottom = pd.concat([top_10, bottom_10])

# Step 4: Filter original DataFrame for these countries
filtered_df = WT_filtered[WT_filtered['country'].isin(top_bottom.index)]

# Step 5: Set plot size and create countplot
plt.figure(figsize=(10, 8))
ax = sns.countplot(data=filtered_df, y='country', order=top_bottom.index)

# Step 6: Add data labels to bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3)

# Step 7: Finalize plot
plt.title('Top 10 and Bottom 10 Countries by Number of Cities')
plt.xlabel('Number of Cities')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Count cities per country
country_counts = WT_filtered['country'].value_counts()

# Step 2: Get top 10 and bottom 10
top_10 = country_counts.head(10)
bottom_10 = country_counts.tail(10)

# --- Plot 1: Top 10 Countries (Horizontal Bar Plot) ---
plt.figure(figsize=(10, 6))
ax1 = sns.barplot(x=top_10.values, y=top_10.index, palette="crest")
ax1.set_title("Top 10 Countries by Number of Cities")
ax1.set_xlabel("Number of Cities")
ax1.set_ylabel("Country")

# Add data labels
for i, v in enumerate(top_10.values):
    ax1.text(v + 0.5, i, str(v), color='black', va='center')

plt.tight_layout()
plt.show()

# --- Plot 2: Bottom 10 Countries (Vertical Bar Plot) ---
plt.figure(figsize=(10, 6))
ax2 = sns.barplot(x=bottom_10.index, y=bottom_10.values, palette="rocket")
ax2.set_title("Bottom 10 Countries by Number of Cities")
ax2.set_ylabel("Number of Cities")
ax2.set_xlabel("Country")
plt.xticks(rotation=45)

# Add data labels
for i, v in enumerate(bottom_10.values):
    ax2.text(i, v + 0.2, str(v), color='black', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Average scores by region
avg_scores = WT_filtered.groupby('region')[['culture', 'adventure', 'nature', 'cuisine']].mean()

# Plot grouped bar chart
avg_scores.plot(kind='bar', figsize=(12,6))
plt.title('Average Category Scores by Region')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 cities by culture
top10_culture = WT_filtered.groupby('city')['culture'].mean().sort_values(ascending=False).head(10)

# Bottom 10 cities
bottom10_culture = WT_filtered.groupby('city')['culture'].mean().sort_values(ascending=True).head(10)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Top 10 ---
top10_culture.plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Top 10 Cities by Culture Score')
axes[0].set_xlabel('Average Culture Score')
axes[0].invert_yaxis()  # Optional: keep highest at top

# Add data labels
for i, v in enumerate(top10_culture):
    axes[0].text(v + 0.05, i, f"{v:.2f}", va='center')

# --- Bottom 10 ---
bottom10_culture.plot(kind='barh', ax=axes[1], color='salmon')
axes[1].set_title('Bottom 10 Cities by Culture Score')
axes[1].set_xlabel('Average Culture Score')
axes[1].invert_yaxis()  # Optional

# Add data labels
for i, v in enumerate(bottom10_culture):
    axes[1].text(v + 0.05, i, f"{v:.2f}", va='center')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the boxplot
ax = sns.boxplot(data=WT_filtered, x='budget_level', y='nature')
plt.title('Distribution of Nature Score by Budget Level')

# Calculate medians
medians = WT_filtered.groupby('budget_level')['nature'].median()

# Add median labels on the boxes
positions = range(len(medians))
for pos, median in zip(positions, medians):
    ax.text(pos, median + 0.1, f'{median:.2f}', ha='center', color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
means = WT_filtered.groupby('budget_level')['nature'].mean()
for pos, mean in zip(positions, means):
    ax.text(pos, mean + 0.1, f'Mean: {mean:.2f}', ha='center', color='blue', fontstyle='italic')

plt.tight_layout
plt.show()
```

```python
counts = WT_filtered['budget_level'].value_counts()
for pos, count in zip(positions, counts.loc[medians.index]):
    ax.text(pos, WT_filtered['nature'].min() - 1, f'n={count}', ha='center', color='gray')

plt.show()
```
